```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch GME stock data using yfinance
GME = yf.Ticker("GME")
GME_data = GME.history(period="1y")  # Fetch data for the last 1 year

# Reset the index to get the 'Date' column
GME_data.reset_index(inplace=True)

# Define the make_graph function
def make_graph(df, title="GME Stock Price Over the Last Year"):
    plt.figure(figsize=(10, 4))
    plt.plot(df['Date'], df['Close'], marker='o', color='r', linestyle='-', label='Closing Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.show()
# Call the make_graph function with GMEa stock data
make_graph(GME_data, title="GME Stock Price Over the Last Year")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch Tesla stock data using yfinance
TESLA = yf.Ticker("TSLA")
tesla_data = TESLA.history(period="1y")  # Fetch data for the last 1 year

# Reset the index to get the 'Date' column
tesla_data.reset_index(inplace=True)

# Define the make_graph function
def make_graph(df, title="Tesla Stock Price Over the Last Year"):
    plt.figure(figsize=(10, 4))
    plt.plot(df['Date'], df['Close'], marker='o', color='r', linestyle='-', label='Closing Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.grid(True)
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Call the make_graph function with Tesla stock data
make_graph(tesla_data, title="Tesla Stock Price Over the Last Year")
```
